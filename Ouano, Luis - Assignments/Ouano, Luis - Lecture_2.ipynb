$\textbf{Question 1:}$ Demonstrate that the expected value $E[y]$ can be represented as $\Sigma y \cdot P(y) = n \cdot p_i$.

<b>Step 1:</b> Rewrite the expected value $E[y]$ using $P(y)$:

$E[y] = \Sigma_{y} \, y \cdot P(y)$

<b>Step 2:</b> Substitute the provided joint probability distribution for $P(y)$:

$E[y] = \Sigma_{y} \, y \cdot (\frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k})$

<b>Step 3:</b> Expand the summation:

$E[y] = \Sigma_{y} \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k)$

<b>Step 4:</b> Rearrange terms:

$E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot \Sigma_{y} (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k)$

<b>Step 5:</b> Recognize that the sum is a product of sums:

$E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot  ( \Sigma_{y_1} \,y_1 \cdot p_1) \cdot (\Sigma_{y_2} \, y_2 \cdot p_2) \cdot ( \Sigma_{y_3} \, y_3 \cdot p_3) \ldots ( \Sigma_{y_k} \, y_k \cdot p_k)$

<b>Step 6:</b> Each individual sum represents the expected value of a binomial distribution with parameters $n$ and $p_i$:

$E[Y_i] = n \cdot p_i$

<b>Step 7:</b> Substitute the expected values of each $Y_i$:

$E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) \ldots  (n \cdot p_k)$

<b>Step 8:</b> Notice that $n!$ and $y_1!y_2!y_3!...y_k!$ cancel out:

$E[y] = n \cdot p_1 \cdot n \cdot p_2 \cdot n \cdot p_3 \ldots n \cdot p_k$

Combining terms simplifies the expression, and recognizing that the coefficient $(y_1!y_2!y_3!...y_k!)$ simplifies to $n!$

<b>Step 9:</b> Simplify:

$E[y] = n  (p_1 \cdot  p_2 \cdot  p_3 \ldots  p_k)$

<b>Step 10:</b> Recall that $p_1 + p_2 + p_3 + \ldots + p_k = 1$:

$E[y] = n \cdot 1$
$E[y] = n \cdot p_i$

<b>Conclusion:</b> Therefore, the expected value $E[y]$ equals $\Sigma y \cdot P(y)$, which is equivalent to $n \cdot p_i$.

$\textbf{Question 3: }$ Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?
<br><br>

<b>Given:</b>

-  Probability of success (recovery) $p = 0.30$
-  Number of trials $n = 10$
- Number of successes (recoveries) $y = 9$

<b>Solution:</b>

Using the binomial probability mass function:

$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$

$P(Y \geq 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} + \binom{10}{10} \cdot (1 - 0.30)^{10-10}$

$P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (0.70)^1 + 1 \cdot (0.30)^{10}  \cdot (0.70)^0$

$ P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot 0.70 + (0.30)^{10}$

$ P(Y \geq 9) = 0.000137781 + 0.0000059049$

$ P (Y \geq 9) \approx 0.0001436869$ or $0.014 \%$

<b>Conclusion: </b>Therefore, the probability that at least nine of ten receiving the medication will recover is $0.014 \%$.

$\textbf{Question 4: }$ Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.
<br><br>

<b>Given:</b>
- Probability of success (a fuse being non-defective) $p = 0.95$
- Number of trials $n = 5$

<b>Solution:</b>

$\textbf{Solution:}$

- To find the probability of observing at least one defective fuse, we can compute the complement of the probability of finding none defective.
- So, $P(\text{At least one defective}) = 1 - P(\text{None defective})$.
- This can also be expressed as $P(\text{At least one defective}) = 1 - P(Y = 0)$.

Using the binomial probability mass function:

$P(Y=0) = \binom {5}{0} \cdot (0.05)^0 \cdot (1 - 0.05)^{5-0}$

$P(Y = 0) = (0.95)^5$

$P(\text{At least one defective}) = 1 - (0.95)^5$

$P(\text{At least one defective}) \approx 1 - 0.7738$

$P(\text{At least one defective}) \approx 0.2262$

$\textbf{Conclusion:}$ Hence, the probability of observing at least one defective fuse in a sample of 5 fuses is approximately $0.2262$ or $22.62\%$.

$\textbf{Question 1: }$ Find a general formula for the expected value of this distribution $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.
<br><br>

Given probability distribution: 

$p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$

Expected value formula: 

$E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot p(y)$

<br>

<b>Step 1: </b>Plug the Probability Distribution into the Expected Value Formula

$E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

<b>Step 2: </b>Factor out Constants

$E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \, \frac{\lambda^y}{(y - 1)!} $

<b>Step 3: </b>Since $y = 0$ renders $\frac{\lambda^y}{(y - 1)!}$ undefined, adjust the summation index to commence from $y = 1$.

$E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \, \frac{\lambda^y}{(y - 1)!} $

<b>Step 4: </b>Rewrite $\frac{\lambda^y}{(y - 1)!}$ as $\frac{\lambda^{(y-1)+ 1}}{(y - 1)!}$ 

$E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \,  \frac{\lambda^{(y-1)+ 1}}{(y - 1)!}$

<b>Step 5: </b>Adjust the summation index to start from $y = 0$ to simplify the expression.

$E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \,  \frac{\lambda^{y+1}}{y!}$

<b>Step 6: </b>Factor out $\lambda$

$E[Y] = e^{-\lambda} \lambda \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}$

<b>Step 7: </b>Recall that $e^{\lambda} =  \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}$

$E[Y] = e^{-\lambda} \lambda e^{\lambda} $

<b>Step 8: </b>Simplify: 

$E[Y] =  \lambda$

<b>Conclusion: </b>To conclude, the general expression for the expected value of the given distribution is $E[Y] =  \lambda$
