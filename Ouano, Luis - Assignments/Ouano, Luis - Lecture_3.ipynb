### $\textbf{Directed Graph (Digraphs)}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 1)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, arrowsize=20)
plt.show()
```

The graph above shows an example of a directed graph. A directed graph, also known as a digraph, is a type of graph in which edges have a direction associated with them. In other words, each edge in a directed graph has an explicit direction from one vertex to another.

### $\textbf{Simple Graphs}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
G.add_node('E')

G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'D')
G.add_edge('D', 'E')
G.add_edge('E', 'A')

nx.draw(G, with_labels=True, node_color='lightblue', node_size=1000, font_size=12, edge_color='gray')
plt.show()
```

The graph above shows an example of a simple graph. A simple graph is a graph in which each edge connects two distinct vertices, and there is at most one edge between any pair of vertices. Additionally, it does not have any loops or parallel edges

### $\textbf{Complete Graphs}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

for i in range(len(nodes)):
    for j in range(i + 1, len(nodes)):
        G.add_edge(nodes[i], nodes[j])

nx.draw(G, with_labels=True, node_color='lightcoral', node_size=1000, font_size=12)
plt.show()
```

The graph above is an example of a complete graph. A complete graph is a simple graph in which each pair of distinct vertices is connected by a unique edge. In other words, every vertex in a complete graph is directly connected to every other vertex. A complete graph is denoted by $K_n = \frac{n(n-1)}{2}$ , where $K =$ number of edges, $n =$ number of vertices. Using the formula given with $n =$ 5, we can solve for the number of edges which is 10.

### $\textbf{Bipartite Graphs}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4], bipartite=0)  
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)

G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (2, 'C'), (3, 'C'), (4, 'A')])

pos = nx.bipartite_layout(G, [1, 2, 3, 4]) 
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000)

plt.show()
```

The graph above is an example of a bipartite graph. A bipartite graph, also known as a bigraph, is a type of graph where the vertices can be divided into two disjoint and independent sets such that no two vertices within the same set are adjacent.

### $\textbf{Subgraphs}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, font_size=12)
plt.title('Graph G')
plt.show()


subgraph_nodes = [1, 2, 3]  
subgraph = G.subgraph(subgraph_nodes)


nx.draw(subgraph, pos, with_labels=True, node_color='lightcoral', node_size=1500, font_size=12)
plt.title('Graph H')
plt.show()

```

In the example above, Graph H is considered a subgraph of Graph G because every vertex in 𝐻 is also a vertex in 𝐺 and every edge in H is also an edge in G.

### $\textbf{Travelling in Graphs}$

#### $\textbf{Walk}$

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 24
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])

edgelist = [(1, 2), (2, 3), (3, 4), (4, 5), (6, 1), (6, 5), (7, 5), (7, 4), (8, 4), (8, 3),(4,6), (2, 4), (7,8)]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed=seed)
walk = [(1, 2), (2,4), (4,3), (3,8), (8,4), (4,7), (7,5)]

nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(G, pos=pos, edgelist=walk, edge_color='red', width=2)

plt.show()
```

<b>Route: </b> $v_1 \to v_2 \to v_4 \to v_3 \to v_8 \to v_4 \to v_7 \to v_5$

The edges in red represent a walk route. We started the walk at vertex 1 and ended at vertex 5. This satisfies as a walk because a walk can have repeated edges and vertices.

### $\textbf{Connectedness in a Graph}$

#### $\textbf{Connected Components}$

```python
import networkx as nx
import matplotlib.pyplot as plt

seed = 2

G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'])
G.add_edges_from([('A', 'B'), ('B', 'D'), ('D', 'F'), ('C', 'F'), ('E', 'A'), ('C', 'D'), ('E', 'B'), ('C', 'E'), ('G', 'H'), ('H', 'I'), ('I', 'G')])
pos = nx.spring_layout(G, seed=seed)

node_color = ['skyblue' if node not in ['G', 'H', 'I'] else 'salmon' for node in G.nodes]
nx.draw(G, pos=pos, with_labels=True, node_color=node_color)

plt.show()
```

There are 2 connected components in this graph. The first is the connection between G, H, I and the second is the connection between A to F. They are both a subgraph to the main graph $G$. The vertices of the first subgraph does not connect to any vertices to the other subgraph. Thus, both of them are connected components.

### $\textbf{Euler's Circuit}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])

edgelist = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'F'), ('F', 'A'), ('F', 'B'), ('F', 'D'), ('B', 'D')]
G.add_edges_from(edgelist)
pos = nx.circular_layout(G)

nx.draw(G, pos=pos, with_labels=True)

plt.show()
```

<b>Route: </b> $A \to B \to C \to D \to E \to F \to D \to B \to F \to A$

The graph and route above shows an example of a Euler's Circuit. A Euler's Circuit uses every vertex at least once and passes through every edge exactly once. 

### $\textbf{Euler's Trail}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G'])

edgelist = [('A','B'),('B','C'),('C','A'),('C','D'),('D','E'),('E','C'),('F','C'),('G','C')]
G.add_edges_from(edgelist)

nx.draw(G, with_labels=True)

plt.show()
```

<b>Route: </b> $ F \to C \to D \to E \to C \to B \to A \to C \to G$
 
The graph and route above IS an example of a Euler's Trail. A graph will contain a Euler's Trail if it contains at most two vertices of odd degree while passing through every vertex at least once and every edge exactly once. The route starts a F and ends with G.

### $\textbf{The Travelling Salesman Problem}$

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = ['A', 'B', 'C', 'D', 'E']
for node in nodes:
    G.add_node(node)

edges = [('A', 'B', 2), ('A', 'C', 3), ('A', 'D', 1), ('A', 'E', 4),
         ('B', 'C', 2), ('B', 'D', 2), ('B', 'E', 3),
         ('C', 'D', 1), ('C', 'E', 2),
         ('D', 'E', 3)]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

def tsp(graph):
    min_distance = float('inf')
    optimal_path = None
    nodes = list(graph.nodes())

    for path in permutations(nodes):
        distance = 0
        for i in range(len(path) - 1):
            distance += graph[path[i]][path[i+1]]['weight']
        distance += graph[path[-1]][path[0]]['weight'] 
        if distance < min_distance:
            min_distance = distance
            optimal_path = path
    return optimal_path, min_distance

optimal_path, min_distance = tsp(G)

pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue')
nx.draw_networkx_edges(G, pos, edge_color='gray')
nx.draw_networkx_edges(G, pos, edgelist=[(optimal_path[i], optimal_path[i+1]) for i in range(len(optimal_path)-1)], edge_color='red', width=2)
nx.draw_networkx_edges(G, pos, edgelist=[(optimal_path[-1], optimal_path[0])], edge_color='red', width=2) 

edge_labels = {(edge[0], edge[1]): edge[2] for edge in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title(f"Travelling Salesman Problem\nOptimal Tour: {optimal_path}\nTotal Distance: {min_distance}")
plt.show()
```
